```python
import pandas as pd
```

```python
data = pd.read_csv('titanic.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data['Fare'].ffill(inplace = True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.groupby('Sex')['Fare'].agg(['mean' , 'median' , 'min' , 'max' , 'std' , 'sum'])
```

```python
def age_group(age):
    if age < 18:
        return 'Under 18'
    elif age < 30:
        return '18-29'
    elif age < 40:
        return '30-39'
    elif age < 50:
        return '40-49'
    elif age < 60:
        return '50-59'
    else:
        return '60+'
```

```python
data['Age Group'] = data['Age'].apply(age_group)
```

```python
data.groupby('Age Group')['Fare'].agg(['mean', 'median', 'min', 'max', 'std'])
```
