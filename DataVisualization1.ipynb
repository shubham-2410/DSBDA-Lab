```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = sns.load_dataset('titanic')
```

```python
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.ffill(inplace=True)
```

```python
data.bfill(inplace = True)
```

```python
data.isnull().sum()
```

```python
data['fare'] = np.log(data['fare'])
```

```python
sns.set_style("whitegrid")
sns.set_context("paper", font_scale=1.2)
sns.histplot(data = data , x= 'fare', kde = True )
```

```python
sns.countplot(data = data , x= 'survived')
```

```python
sns.countplot(data = data , x= 'survived' , hue = 'sex')
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data = data , x= 'survived' , hue='age' , palette = 'hot' )
```

```python
df = data['sex'].value_counts()
plt.pie(x=df, autopct = '%1.1f%%' )
plt.show()
```

```python
df
```

```python
sns.scatterplot(data=data, x='age', y='fare', hue='survived')
plt.title('Relationship between Age and Fare, colored by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
